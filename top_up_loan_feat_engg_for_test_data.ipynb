```
col = ['AGE', 'MonthlyIncome']
```

```
def impute_nan(test, feat):
    m=test[feat].median()
    test[feat]=test[feat].fillna(m)

for i in col:
    impute_nan(test, i)
```

```
test['Area'] = test['Area'].fillna('Other')
test['SEX'] = test['SEX'].fillna('M')
test['City'] = test['City'].fillna('Other')
test['MaturityDAte'] = test['MaturityDAte'].fillna('2018-06-10')
```

```
import datetime as dt
test[["MaturityDAte", "DisbursalDate", "AuthDate"]] = test[["MaturityDAte", "DisbursalDate", "AuthDate"]].apply(pd.to_datetime)
```
