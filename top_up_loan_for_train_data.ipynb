```
train_b['INSTALLMENT-AMT']=train_b['INSTALLMENT-AMT'].str.replace(',', '')
train_b['CREDIT-LIMIT/SANC AMT']=train_b['CREDIT-LIMIT/SANC AMT'].str.replace(',', '')
train_b['DISBURSED-AMT/HIGH CREDIT']=train_b['DISBURSED-AMT/HIGH CREDIT'].str.replace(',', '')
train_b['CURRENT-BAL']=train_b['CURRENT-BAL'].str.replace(',', '')
train_b['OVERDUE-AMT']=train_b['OVERDUE-AMT'].str.replace(',', '')
train_b.head()
```

```
train_b['INSTALLMENT-AMT'].value_counts()
```

```
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace(r'Single Payment Loan \(bullet / balloon\)', 'Monthly')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Monthly', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Other', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Quarterly', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/BiWeekly', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Weekly', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Annually', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/BiMonthly', '')
train_b['INSTALLMENT-AMT'] = train_b['INSTALLMENT-AMT'].str.replace('/Semi annually', '')
```

```
train_b.head()
```

```
train_b['CREDIT-LIMIT/SANC AMT'] = pd.to_numeric(train_b['CREDIT-LIMIT/SANC AMT'])
train_b['DISBURSED-AMT/HIGH CREDIT'] = pd.to_numeric(train_b['DISBURSED-AMT/HIGH CREDIT'])
train_b['INSTALLMENT-AMT'] = pd.to_numeric(train_b['INSTALLMENT-AMT'])
train_b['CURRENT-BAL'] = pd.to_numeric(train_b['CURRENT-BAL'])
train_b['OVERDUE-AMT'] = pd.to_numeric(train_b['OVERDUE-AMT'])
train_b['SELF-INDICATOR'] = train_b['SELF-INDICATOR'].astype(object)
```

```
train_b.head()
```

```
train_b.info()
```

```
train_b.fillna(0, inplace=True)
train_b.head()
```

```
agg_func = {'CREDIT-LIMIT/SANC AMT' : ['mean'], 'DISBURSED-AMT/HIGH CREDIT' : ['mean'], 'INSTALLMENT-AMT': ['mean'], 'CURRENT-BAL': ['mean'], 'OVERDUE-AMT': ['mean'], 'WRITE-OFF-AMT': ['mean'], 'TENURE': ['mean']}
agg_name = train_b.groupby(['ID']).agg(agg_func)
agg_name.columns = [ 'AGG_' + ('_'.join(col).strip()) for col in agg_name.columns.values]
agg_name.reset_index(inplace=True)
```
